```python
# Dependencies

import pandas as pd
import numpy as np
import glob
import os
import timeit                                # To check performance
```

```python
# Export clean CSV file to master CSV file

tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df = pd.read_csv("data_set_exports/ecobici_ride_data-oneInHundred.csv", index_col=None, header=0)

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to export unified CSV file: {round(toc - tic, 2)}')
```

```python
ecobici_rides_df.info()
```

```python
# Caste date columns as datetime

tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df['Fecha_Retiro'] = pd.to_datetime(ecobici_rides_df['Fecha_Retiro'])
ecobici_rides_df['Fecha_Arribo'] = pd.to_datetime(ecobici_rides_df['Fecha_Arribo'])

toc = timeit.default_timer() 
print(f'Processing time (in seconds): {round(toc - tic, 2)}')

ecobici_rides_df
```

```python
ecobici_rides_df_test = ecobici_rides_df.dropna().copy()
ecobici_rides_df_test.info()
```

```python
tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df_test['Hora_Retiro'] = pd.to_datetime(ecobici_rides_df_test['Hora_Retiro']).dt.time
ecobici_rides_df_test['Hora_Arribo'] = pd.to_datetime(ecobici_rides_df_test['Hora_Arribo']).dt.time

toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to process data: {round(toc - tic, 2)}')

ecobici_rides_df_test
```

```python
ecobici_rides_df_test['Hora_Retiro']
```

```python
tic = timeit.default_timer()                            # Monitor performance

ecobici_rides_df_test['Hora_Arribo'] = pd.to_datetime(ecobici_rides_df_test['Hora_Arribo']).dt.time


toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to process data: {round(toc - tic, 2)}')

ecobici_rides_df_test
```

```python
# Combine Fecha_Retiro and Hora_Retiro string and convert to datetime

tic = timeit.default_timer()                            # Monitor performance

for index, row in ecobici_rides_df.iterrows():
               
    try: 
        ecobici_rides_df_test['Hora_Retiro'] = pd.to_datetime(ecobici_rides_df_test['Hora_Retiro'])
        ecobici_rides_df_test['Hora_Arribo'] = pd.to_datetime(ecobici_rides_df_test['Hora_Arribo'])
        if index%1000000 == 0: print(f'We\'re up to {index}!')
               
    except: 
        print(f'ERROR: ROW {index} DROPPED')
        ecobici_rides_df_test.drop(index, inplace=True, axis=0)
                                          
toc = timeit.default_timer()                            # Monitor performance
print(f'Time (in seconds) to process data: {round(toc - tic, 2)}')

ecobici_rides_df_test
```
